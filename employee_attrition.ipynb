# Employee Attrition Predictor

## Objective
Predict whether an employee is likely to leave the company using HR data,
and identify the key factors that drive attrition.

## Dataset
IBM HR Analytics Employee Attrition Dataset (available on Kaggle)
- 1470 employee records
- 35 features including job role, satisfaction scores, overtime, salary and more

## Approach
- Exploratory Data Analysis (EDA)
- Feature Engineering and Encoding
- Handled class imbalance using SMOTE
- Compared 3 models: Logistic Regression, Random Forest, XGBoost
- Best Model: Logistic Regression (ROC-AUC: 0.79, Class 1 F1: 0.49)

## Key Findings
- Overtime is the strongest predictor of employee attrition
- Frequent business travel and lack of promotions significantly increase attrition risk
- Employees with longer tenure and stable manager relationships are more likely to stay
- Model can help HR teams proactively identify at-risk employees

```python
df.drop(['EmployeeCount', 'EmployeeNumber'], axis = 1, inplace = True)
df.head()
```

```python
df.drop(['Over18', 'StandardHours'], axis = 1, inplace = True)
df.head()
```

```python
sns.countplot(x = "Attrition", hue = "YearsWithCurrManager", data = df)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['Attrition', 'OverTime', 'Gender']:
  df[col] = le.fit_transform(df[col])
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
df = pd.get_dummies(df, columns = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'], drop_first = True)
df.head()
```

```python
X = df.drop('Attrition', axis = 1)
Y = df['Attrition']
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
y_pred = lr.predict(X_test)
print(accuracy_score(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```

```python
print("ROC-AUC:", roc_auc_score(Y_test, lr.predict_proba(X_test)[:,1]))
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42)
X_train_sm, Y_train_sm = sm.fit_resample(X_train, Y_train)

print("Before SMOTE:",Y_train.value_counts().to_dict())
print("After SMOTE:" ,Y_train_sm.value_counts().to_dict())
```

```python
from sklearn.ensemble import RandomForestClassifier
rf2 = RandomForestClassifier(n_estimators=100, random_state=42)
rf2.fit(X_train_sm, Y_train_sm)

y_pred_rf2 = rf2.predict(X_test)

print(classification_report(Y_test, y_pred_rf2))
print("ROC-AUC:", roc_auc_score(Y_test, rf2.predict_proba(X_test)[:,1]))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred_rf2))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(scale_pos_weight = 6.5, random_state = 42, eval_metric = 'logloss')
xgb.fit(X_train_sm, Y_train_sm)
y_pred_xgb = xgb.predict(X_test)

print(classification_report(Y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(Y_test, xgb.predict_proba(X_test)[:,1]))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred_xgb))
```

```python
lr2 = LogisticRegression(random_state = 42, max_iter = 1000)
lr2.fit(X_train_sm, Y_train_sm)
y_pred_lr2 = lr2.predict(X_test)

print(classification_report(Y_test, y_pred_lr2))
print("ROC-AUC:", roc_auc_score(Y_test, lr2.predict_proba(X_test)[:,1]))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred_lr2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Get feature names
feature_names = df.drop('Attrition', axis=1).columns

# Get coefficients
coefficients = pd.Series(lr.coef_[0], index=feature_names)

# Sort and plot top 15
top_features = coefficients.abs().sort_values(ascending=False).head(15)

plt.figure(figsize=(10,6))
coefficients[top_features.index].sort_values().plot(kind='barh', color='steelblue')
plt.title('Top 15 Features Influencing Attrition')
plt.xlabel('Coefficient Value')
plt.tight_layout()
plt.show()
```
